```python
import numpy as np
import math
```

```python
class LinearRegression:
    def __init__(self, X, y, lr=0.001, benchmark=0.00001):
        self.X = X
        self.y = y
        self.lr = lr
        self.benchmark = benchmark
        
        self.w = np.random.randn(self.X.shape[1])
    
    def predict(self, X):
        return np.dot(X, self.w)
    
    def error(self, X, y):
        return abs((self.predict(X) - y));
    
    def calc_grad(self, X, y):
        
        grads = np.dot(X.T, self.error(self.X, self.y))

        return grads
    
    def update_weights(self, grads):
        
        for i in range(len(self.w)):
            self.w[i] -= self.lr * grads[i]
    
    def train(self):
        min_error = self.error(self.X, self.y).mean();
        
        cnt_err_bigger_than_pred = 0;
        
        while(True):
            self.update_weights(self.calc_grad(self.X, self.y))
            error = self.error(self.X, self.y).mean()
            
            if(min_error - error < self.benchmark): 
                self.lr -= 0.00001;
                cnt_err_bigger_than_pred += 1;
            else:
                min_error = error;
                cnt_err_bigger_than_pred = 0;
            
            if(cnt_err_bigger_than_pred > 10): break
            
        print("Min Error: ", min_error);
        
```

```python
from sklearn.datasets import make_regression
X, y, coef = make_regression(
    n_samples=1000,
    n_features=1,
    n_informative=1,
    noise=10,
    coef=True,
    random_state=0,
)
```

```python
import matplotlib.pyplot as plt

plt.scatter(X, y)
```

```python
print("tr X: ", X)
print("tr y: ", y)
```

```python
model = LinearRegression(X=X, y=y, lr=0.001, benchmark=0.001)
```

```python
model.train()
```

```python
pred = model.predict(X)

plt.scatter(X, y)
plt.scatter(X, pred, marker='.', color='red')
```
